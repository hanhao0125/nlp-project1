```python
import  pandas as pd
from stanfordcorenlp import StanfordCoreNLP
from gensim.models import KeyedVectors
import random
from pprint import pprint
base_path = '/root/files/'
file_path = base_path + 'news_content.csv'
nlp_path = '/root/stanford-corenlp'
word2vec_model_path = base_path + 'sgns.merge.word.bz2'
```

```python
wm = KeyedVectors.load_word2vec_format(word2vec_model_path,
                                       binary=False, 
                                       unicode_errors='ignore')
```

```python
a = wm.most_similar('说',topn=20)
for i in a: 
    print(i)
say_represents = set(a)
```

# 读取新闻

```python
d = pd.read_csv(file_path)
```

```python
def process_content(c:pd.DataFrame) -> pd.DataFrame:
    c['content'] = c['content'].apply(lambda x:str(x).replace('\n',' '))
    return c
```

```python
d = process_content(d)
print(d.head())
```

```python
nlp = StanfordCoreNLP(nlp_path,lang='zh')
```

```python
sentences = d.content.tolist()
sentences[0]
```

```python
s = random.randint(1,10000)
sentence = sentences[s]
sentence
```

```python
print(nlp.word_tokenize(sentence))
print(nlp.pos_tag(sentence))
print(nlp.ner(sentence))
for i in nlp.ner(sentence):
    print(i)
# nlp.close()
# print(nlp.parse(sentence))
# print(nlp.dependency_parse(sentence))
```

```python
names = {'PERSON','ORGANIZATION'}
end = {'。','!','?'}
print(say_represents)
def find(s:str):
    points = []
    
    ners = nlp.ner(s)
    words = [i[0] for i in ners]
    k = 0
    print(''.join(words))
    while k < len(ners):
        if ners[k][1] in names:
            _p = [ners[k][0]]
            start = k
            while k < len(ners) and words[k] not in end: k+=1
            _p.append(''.join(words[start:k]))
            points.append(_p)
        k += 1
    pprint(points)

find(sentence)
```
